```python
import numpy as np
import matplotlib.pyplot as plt

# Sıradan verilən yaratma (NaN verilənləri daxil olması şərti ilə)
x = np.array([1, 2, 3, 4, np.nan, 6, 7, np.nan,16,19,15,18,22])
y = np.array([3, 5, np.nan, 4, 2, 6, np.nan, 8,9,12,15,19,22])

# line plot yaratma
plt.plot(x, y, marker='o')


# koordinatlar
plt.xlabel('X')
plt.ylabel('Y')

# Başlıq
plt.title('NaN verilənlərin line qrafiki')

# Qrafikin göstərilməsi
plt.show()
```

```python
import numpy as np
import seaborn as sns
# Sıradan verilən yaratma (NaN verilənləri daxil olması şərti ilə)
x = np.array([1, 2, 3, 4, np.nan, 6, 7, np.nan,16,19,15,18,22])
y = np.array([3, 5, np.nan, 4, 2, 6, np.nan, 8,9,12,15,19,22])

#Np stack yaratma
data = np.array([x, y])

#Heatmap-ilə vizualizasiyası
sns.heatmap(data, cmap='viridis', cbar=True)
plt.show()
```
