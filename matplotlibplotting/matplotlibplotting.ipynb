```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,50,30)
y = np.sin(x) + 0.1 *np.random.randint(len(x))
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,"--")
```

```python
plt.plot(x,y,'o--')
```

```python
plt.plot(x,y,"o--",color = "red")
```

Change to Marker size(ms), Line with(lw)

```python
plt.plot(x,y,lw = 0.15,ms = 0.01)  
```

```python
plt.figure(figsize=(10,8))
plt.plot(x,y,"o--",color = "red",lw = 2)
```

```python
plt.figure(figsize = (10,3))
plt.xlabel("Time")
plt.ylabel("Electron spin")
plt.plot(x,y,"o--" ,color = "purple",lw = 1,label = "data")
plt.legend(ncols = 1,loc = "upper right")
```

```python
omega1 = 0.2+np.random.randn(1000)*0.2+0.4
omega2 = 0.25+np.random.randn(1000)*0.2+0.4
omega3 = 0.2+np.random.randn(1000)+0.4
omega4 = 0.25+np.random.randn(1000)*0.5+0.4
```

```python
fig,axis =plt.subplots(2,2,figsize = (12,12))
ax = axis[0][0]
plt.hist(omega1, bins = 30, color= "red", density= True,label= "omega1")
plt.legend(loc = "upper right")
plt.text(0.5,0.81,"omega1")
plt.show()
ax = axis[1][0]
plt.hist(omega2, bins = 30, color= "green", density= True,label= "omega1")
plt.legend(loc = "upper right")
plt.text(0.5,0.81,"omega2")
plt.show()
ax = axis[1][1]
plt.hist(omega3, bins = 30, color= "blue", density= True,label= "omega1")
plt.legend(loc = "upper right")
plt.text(0.5,0.81,"omega3")
plt.show()
ax = axis[0][1]
plt.hist(omega4, bins = 30, color= "purple", density= True,label= "omega1")
plt.legend(loc = "upper right")
plt.text(0.5,0.81,"omega4")
plt.show()
```

2D plot

```python
l = np.linspace(-1,1,100)
x,y = np.meshgrid(l,l)
z = x**2+x*y
```

```python
z
```

```python
x
```

```python
plt.contourf(x,y,z)
```

```python
c = plt.contour(x,y,z, cmap = "plasma")
plt.clabel(c, fontsize = 8)
plt.colorbar(c)
```

```python
fig2 ,ax2 = plt.subplots(subplot_kw={"projection": "3d"})
ax2.plot_surface(x,y,z, cmap = "plasma")
```

Animation

```python
from matplotlib import animation
from matplotlib.animation import PillowWriter
fig3 ,ax3 = plt.subplots(subplot_kw={"projection": "3d"})
ax3.plot_surface(x,y,z, cmap = "coolwarm")
ax3.view_init(elev = 10,azim = 0)

def animate(i):
    ax3.view_init(elev = 10,azim = 3*i)


ani = animation.FuncAnimation(fig3,animate,frames= 50,interval = 150)
ani.save("pillow2frame50fps10interval450.gif",writer="pillow",fps = 50,dpi = 100)
```
