```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
import statsmodels.api as sm
```

```python
retail_data = pd.read_csv("/var/home/rafo/vscode02/decomposition_additive model/RetailTurnover.csv")
```

```python
retail_data["Year"] = pd.to_datetime(retail_data["Year"],format="%Y")
```

```python
retail_data
```

```python
fig = plt.gca()
fig.plot(retail_data["Year"],retail_data["Turnover"])
plt.xlabel("Turnover")
plt.ylabel("Date")
plt.legend(loc = "best")
```

Decompose the time series

```python
date_range = pd.date_range(start='1/7/1982', end='31/3/1992',
freq='Q')
retail_data['TimeIndex'] = pd.DataFrame(date_range,
columns=['Quarter'])
```

```python
decomp_retil_data = seasonal_decompose(retail_data["Turnover"],model= "additive",period = 4)
```

```python
decomp_retil_data.plot()
plt.show()
```

```python
trend= decomp_retil_data.trend
seasonal = decomp_retil_data.seasonal
residual = decomp_retil_data.resid
```
