```python
import seaborn as sns

# Datasetin yüklənməsi
mpg = sns.load_dataset("mpg")

# Əlaqəli dəyişənlərə görə müqayisəli qrafik
sns.relplot(x="horsepower",
            y="mpg", hue="origin",
            data=mpg).set( title = "Numeric");
```

```python
import seaborn as sns

# Dataseti yüklənməsi
tips = sns.load_dataset("tips")

# Kategorik dəyişənlərə görə müqayisəıi qrafik
sns.catplot(x="day",
            y="total_bill",
            hue="sex",
            data=tips,
            kind="bar").set(title = "Kategorik");
```
